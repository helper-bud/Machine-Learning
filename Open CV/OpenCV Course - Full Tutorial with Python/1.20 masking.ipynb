```python
import numpy as np
from google.colab.patches import cv2_imshow # needed for collab
import cv2
```

```python
imr = cv2.imread("/content/cats 2.jpg") # importing image
cv2_imshow(imr) # showing image
```

```python
img = cv2.resize(imr,(150,150)) #(which image,(width, height))
cv2_imshow(img)
```

**Masking**

Masking in open cv is essentially allow us to focus on certain parts of an image. 

For example , if you are interested in focusing on peoples faces from a photo, you can essentially apply masking  remove all the unwanted parts of the image. 


```python
blank = np.zeros(img.shape[:2], dtype='uint8') #the dimension of the mask have to be the same size as of the image.
cv2_imshow(blank)
```

```python
circle = cv2.circle(blank.copy(), (img.shape[1]//2,img.shape[0]//2), 100, 255, -1)
cv2_imshow(circle) # this is actully our mask 
```

```python
rectangle = cv2.rectangle(blank.copy(), (30,30), (370,370), 255, -1)
weird_shape = cv2.bitwise_and(circle,rectangle)
cv2_imshow(weird_shape)
```

```python
masked = cv2.bitwise_and(img,img,mask=weird_shape)
cv2_imshow( masked)
```

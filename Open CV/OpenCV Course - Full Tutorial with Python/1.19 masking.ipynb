**BitWisw Operator**

Basic bitwise operators are and or, xor, not

```python
blank = np.zeros((400,400), dtype='uint8')
```

```python
rectangle = cv2.rectangle(blank.copy(), (30,30), (370,370), 255, -1)
cv2_imshow(rectangle)
```

```python
circle = cv2.circle(blank.copy(), (200,200), 200, 255, -1)
cv2_imshow(circle)
```

**bitwise AND**

what basicllay it did, it took two image, placed them over one another and basically returned intersection

```python
bitwise_and = cv2.bitwise_and(rectangle, circle)
cv2_imshow(bitwise_and)
```

**bitwise OR**

non-intersecting and intersecting regions

It also took those images and put them over each other and found the common and not common region of one of these images and basically superimpose them over one another.

```python
bitwise_or = cv2.bitwise_or(rectangle, circle)
cv2_imshow(bitwise_or)
```

**bitwise XOR**

non-intersecting regions

```python
# bitwise XOR --> non-intersecting regions
bitwise_xor = cv2.bitwise_xor(rectangle, circle)
cv2_imshow(bitwise_xor)
```

**bitwise NOT**

It basically inverst the color

```python
# bitwise NOT
bitwise_not = cv2.bitwise_not(circle) #  inverted the color of the bitwise not
cv2_imshow(bitwise_not)
```

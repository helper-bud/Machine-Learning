```python
import numpy as np
from google.colab.patches import cv2_imshow # needed for collab
import cv2
import matplotlib.pyplot as plt
```

```python
imr = cv2.imread("/content/lena.jpg") # importing image
cv2_imshow(imr) # showing image
```

```python
img = cv2.resize(imr,(150,150)) #(which image,(width, height))
cv2_imshow(img)
```

**Edge Detection**

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

Lap lacian

```python
lap =cv2.Laplacian(gray, cv2.CV_64F)
lap = np.uint8(np.absolute(lap))
cv2_imshow(lap)
```

**sobel**

Sobel computes the gradients in two directins the x and the y. 

```python
sobelx = cv2.Sobel(gray, cv2.CV_64F,1,0)
sobely = cv2.Sobel(gray, cv2.CV_64F,0,1)
```

```python
cv2_imshow(sobelx)
cv2_imshow(sobely)
```

```python
combined = cv2.bitwise_or(sobelx, sobely)
cv2_imshow(combined)
```

Diffrence with canny image detector. 

```python
canny = cv2.Canny(gray, 150,175)
cv2_imshow(canny)
```

canny is more cleared, that is in most cases you see canny images.

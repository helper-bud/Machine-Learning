```python
import numpy as np
from google.colab.patches import cv2_imshow # needed for collab
import cv2
```

```python
imr = cv2.imread("/content/lon.jpg") # importing image
cv2_imshow(imr) # showing image
```

```python
img = cv2.resize(imr,(150,150)) #(which image,(width, height))
cv2_imshow(img)
```

Color Channels

A color image is basically consist of multiple channel red, green and blue. All the images are basically 3 color channels.

```python
b,g,r = cv2.split(img)
```

```python
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)
```

```python
import numpy as np
from google.colab.patches import cv2_imshow # needed for collab
import cv2
```

```python
imr = cv2.imread("/content/lon.jpg") # importing image
cv2_imshow(imr) # showing image
```

```python
img = cv2.resize(imr,(150,150)) #(which image,(width, height))
cv2_imshow(img)
```

BGR TO GRAY SCALE

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

BGR to HSV (HSV is basically how people see color)


```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv)
```

BGR to L*a*b(more tuned to how human perieved color)

```python
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
cv2_imshow(lab)
```

BGR to RGB

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(rgb)
```

HSV to BGR

```python
lab_bgr = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)
cv2_imshow(lab_bgr)
```

```python
import numpy as np
from google.colab.patches import cv2_imshow # needed for collab
import cv2
```

```python
imr = cv2.imread("/content/lon.jpg") # importing image
cv2_imshow(imr) # showing image
```

```python
img = cv2.resize(imr,(150,150)) #(which image,(width, height))
cv2_imshow(img)
```

Blurring

Averaging:Finds averaging of the surrounding pexel

```python
# Averaging 
average = cv2.blur(img, (3,3))
cv2_imshow( average)
```

Gaussian Blur

```python
gauss = cv2.GaussianBlur(img, (3,3), 0)
cv2_imshow(gauss)
```

Median : Finds the median of the surronding pexels

```python
median = cv2.medianBlur(img, 3)
cv2_imshow(median)
```

Bilateral blur

It will blur the image but will retain the image

```python

bilateral = cv2.bilateralFilter(img, 10, 35, 25)
cv2_imshow(bilateral)
```

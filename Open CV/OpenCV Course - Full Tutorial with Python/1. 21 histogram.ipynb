```python
import numpy as np
from google.colab.patches import cv2_imshow # needed for collab
import cv2
import matplotlib.pyplot as plt
```

```python
imr = cv2.imread("/content/lena.jpg") # importing image
cv2_imshow(imr) # showing image
```

```python
img = cv2.resize(imr,(150,150)) #(which image,(width, height))
cv2_imshow(img)
```

**Histogram**

Histogram alows you to visualize the distribuition of pixel intesities in an image. You can visualize pixel intesities distribution. 

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
cv2_imshow(gray)
```

Gray scale histogram

```python
gray_hist = cv2.calcHist([gray], [0], None, [256], [0,256] ) 
#(images, channels, mask, histsize(basically the nuber of bins that we want to use for computing the histogram), ranges for all possible pixel values)
```

```python
plt.plot(gray_hist) # this will show the distribution of pixel in the image
```

```python
plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
plt.plot(gray_hist)
plt.xlim([0,256])
plt.show()
```

```python
plt.figure()
plt.title('Colour Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
colors = ('b', 'g', 'r')
for i,col in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])

plt.show()
# this plot actullay shows the color channel histogram. 
```

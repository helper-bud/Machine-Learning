```python
import numpy as np
from google.colab.patches import cv2_imshow # needed for collab
import cv2
import matplotlib.pyplot as plt
```

```python
imr = cv2.imread("/content/lena.jpg") # importing image
cv2_imshow(imr) # showing image
```

```python
img = cv2.resize(imr,(150,150)) #(which image,(width, height))
cv2_imshow(img)
```

**Thresholding**

Thresholding is a binary relisation of an image 

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

Simple thresholding

```python
threshold, thresh = cv2.threshold(gray, 150,255, cv2.THRESH_BINARY)
#(image, threshold value, maximum value, thresh  )
cv2_imshow(thresh)
```

the more you decrese the value the more white parts there are.. Look at the photo below

```python
threshold, thresh = cv2.threshold(gray, 100,255, cv2.THRESH_BINARY) # value 100
#(image, threshold value, maximum value, thresh  )
cv2_imshow(thresh)
```

```python
threshold, thresh = cv2.threshold(gray, 50,255, cv2.THRESH_BINARY) # value 50
#(image, threshold value, maximum value, thresh  )
cv2_imshow(thresh)
```

Inverse thresholding

here we inversed the threshold image >  cv2.THRESH_BINARY_INV with this. 

```python
threshold, thresh_inv = cv2.threshold(gray, 150,255, cv2.THRESH_BINARY_INV) # value 50
#(image, threshold value, maximum value, thresh  )
cv2_imshow(thresh_inv)
```

**Adaptive threshold**

you can see from the above we have to give the value manually to threhold. but not every time it works . To let the computer find the optimal thresholded values, we use adaptive threshold.
Using the value it finds , it binarize over the image. 

```python
adaptive_threshold = cv2.adaptiveThreshold(gray,255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 3)
cv2_imshow(adaptive_threshold)
# (source image, max value, adaptive method, threshold type, blcoksize, c, [dst]])
# c value is basiclaly subtracts from the mean value and aloowing us to essenstially fine tune our threshold. 
```

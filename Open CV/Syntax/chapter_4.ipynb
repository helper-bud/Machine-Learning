```python
import numpy as np
from google.colab.patches import cv2_imshow # needed for collab
import cv2
```

Reading Image

```python
img = np.zeros((512,512,3), np.uint8) # (width, height)0 means black. Provides a black image.
```

coloring image

```python
img[200:300] = 255, 0, 0  # blue image [height : width]. dimension still will be 512, 512
cv2_imshow(img)
```

Line creating

```python
cv2.line(img,(0,0), (300, 300),(0,255,0),3) #cv2.line(img,(starting point), (ending point),(color code), thick ness)
cv2_imshow(img)
```

```python
# to bring line from on end to another
cv2.line(img,(0,0), (img.shape[1], img.shape[0]),(0,255,0),3) #cv2.line(img,(widht, height), (ending point),(color code), thick ness)
cv2_imshow(img)
```

Rectangle

```python
# rectangle
cv2.rectangle(img, (0,0), (250,350),(0,0,255), 3)#cv2.rectangle(img,(starting point), (ending point),(color code), thick ness)
cv2_imshow(img)
```

```python
# to fit the middle
cv2.rectangle(img, (0,0), (250,350),(0,0,255), cv2.FILLED)
cv2_imshow(img)
```

circle

```python
cv2.circle(img, (400,50),30,(255,255,0),5)#(img,(center),radius ,color, thickness)
cv2_imshow(img)
```

text

```python
cv2.circle(img, (400,50),30,(255,255,0),5)#(img,(center),radius ,color, thickness)
cv2_imshow(img)
```

text on image

```python
cv2.putText(img, "OPECCV" , (300, 200), cv2.FONT_HERSHEY_SIMPLEX, 1 ,(0,150,0) ) #(img, text, origin, font, scale[increse size], color
cv2_imshow(img)
```

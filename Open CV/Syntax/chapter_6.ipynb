```python
import numpy as np
from google.colab.patches import cv2_imshow # needed for collab
import cv2
```

```python
img = cv2.imread("/content/test.jpg")
cv2_imshow(img)
```

Horizontal Image

```python
image_horizontal = np.hstack((img,img))
cv2_imshow(image_horizontal)
```

Vertical Image

```python
image_vertical = np.vstack((img,img,img)) # depends on how many image you want to sho
cv2_imshow(image_vertical)
```
